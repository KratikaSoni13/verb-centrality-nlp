```
from collections import Counter

LANGS = {
    "English": "/content/data/english/en_ewt-ud-train.conllu",
    "Hindi":   "/content/data/hindi/hi_hdtb-ud-train.conllu",
    "French":  "/content/data/french/fr_gsd-ud-train.conllu",
    "German":  "/content/data/german/de_gsd-ud-train.conllu",
}

results = []
for lang, path in LANGS.items():
    with open(path, "r", encoding="utf-8") as f:
        sents = conllu.parse(f.read())
    pos_counter = Counter()
    total_tokens = 0
    for sent in sents:
        for tok in sent:
            if isinstance(tok["id"], int) and tok["upos"]:
                pos_counter[tok["upos"]] += 1
                total_tokens += 1
    results.append({
        "Language": lang,
        "Sentences": len(sents),
        "Tokens": total_tokens,
        "VERB%": round(100*pos_counter["VERB"]/total_tokens, 1),
        "NOUN%": round(100*pos_counter["NOUN"]/total_tokens, 1),
        "ADJ%":  round(100*pos_counter["ADJ"]/total_tokens, 1),
        "ADV%":  round(100*pos_counter["ADV"]/total_tokens, 1),
    })
```

```

def extract_tokens(sentence, sent_id):
    """
    Extract valid tokens from one CoNLL-U sentence.
    Returns a list of dictionaries, one per valid token.
    """
    tokens = []
    sent_len = sum(
        1 for t in sentence
        if isinstance(t["id"], int) and t["upos"] is not None
    )
    for token in sentence:
        # Skip multi-word tokens (id like '1-2')
        if not isinstance(token["id"], int):
            continue
        # Skip tokens with missing UPOS or HEAD
        if token["upos"] is None or token["head"] is None:
            continue
        # Skip punctuation tokens
        if token["upos"] == "PUNCT":
            continue

        tokens.append({
            "sent_id":  sent_id,
            "token_id": token["id"],
            "form":     token["form"],
            "upos":     token["upos"],
            "head":     token["head"],
            "deprel":   token["deprel"],
            "sent_len": sent_len
        })
    return tokens

    #We remove PUNCT tokens because punctuation has no meaningful syntactic role in centrality analysis.
    #Keeping them would artificially inflate degree centrality of nearby words.
```

```
def load_corpus(filepath, language):

    print(f"Loading {language}...")
    with open(filepath, "r", encoding="utf-8") as f:
        raw = f.read()

    sentences = conllu.parse(raw)
    print(f"  Found {len(sentences)} sentences")

    all_tokens = []
    for i, sent in enumerate(sentences):
        sent_id = sent.metadata.get("sent_id", str(i))
        tokens = extract_tokens(sent, sent_id)
        all_tokens.extend(tokens)

    df = pd.DataFrame(all_tokens)
    print(f"  Extracted {len(df)} valid tokens")
    print(f"  Unique sentences: {df['sent_id'].nunique()}")
    return df
```

```
def sentence_to_graph(sent_df):

    G = nx.DiGraph()

    for _, row in sent_df.iterrows():
        G.add_node(
            int(row["token_id"]),
            form      = str(row["form"]),
            upos      = str(row["upos"]),
            pos_group = str(row["pos_group"]),
            is_root   = bool(row["is_root"])
        )

    for _, row in sent_df.iterrows():
        head = int(row["head"])
        dep  = int(row["token_id"])

        if head == 0:
            continue

        if head in G.nodes:
            G.add_edge(
                head, dep,
                deprel = str(row["deprel"])
            )

    return G
```

```
import pickle
def build_corpus_graphs(csv_path, language):
    """
    Build one NetworkX DiGraph per sentence for an entire corpus.

    Returns:
      graphs: list of (sent_id, DiGraph) tuples
    """
    print(f"Building graphs for {language}...")
    df = pd.read_csv(csv_path)
    df = normalize_pos(df)
    graphs = []
    grouped = df.groupby("sent_id", sort=False)
    total = len(grouped)

    for i, (sent_id, sent_df) in enumerate(grouped):
        G = sentence_to_graph(sent_df)
        graphs.append((sent_id, G))

        # Print progress every 2000 sentences
        if (i + 1) % 2000 == 0:
            print(f"  {i+1}/{total} sentences processed...")

    print(f"  Done. Built {len(graphs)} graphs.")
    return graphs


LANGUAGES = {
    "english": "/content/data/clean_english.csv",
    "hindi":   "/content/data/clean_hindi.csv",
    "french":  "/content/data/clean_french.csv",
    "german":  "/content/data/clean_german.csv",
}

if __name__ == "__main__":
    import os
    os.makedirs("graphs", exist_ok=True)

    for lang, path in LANGUAGES.items():
        graphs = build_corpus_graphs(path, lang)
        out_path = f"graphs/{lang}_graphs.pkl"
        with open(out_path, "wb") as f:
            pickle.dump(graphs, f)
        print(f"  Saved {len(graphs)} graphs to {out_path}\n")
```

```
def validate_graph(G, sent_id):
    """
    Check 4 tree properties. Returns list of error strings.
    Empty list = valid tree.
    """
    errors = []
    N = G.number_of_nodes()
    E = G.number_of_edges()

    if E != N - 1:
        errors.append(f"Expected {N-1} edges, got {E}")

    roots = [n for n in G.nodes() if G.in_degree(n) == 0]
    if len(roots) != 1:
        errors.append(f"Expected 1 root, found {len(roots)}: {roots}")

    if not nx.is_weakly_connected(G):
        errors.append("Graph is not connected")

    if nx.number_of_selfloops(G) > 0:
        errors.append("Self-loop detected")

    return errors


for lang in ["english", "hindi", "french", "german"]:
    with open(f"graphs/{lang}_graphs.pkl", "rb") as f:
        graphs = pickle.load(f)

    invalid = 0
    for sent_id, G in graphs:
        errs = validate_graph(G, sent_id)
        if errs:
            invalid += 1
            if invalid <= 3:
                print(f"[{lang}] {sent_id}: {errs}")

    valid = len(graphs) - invalid
    print(f"{lang}: {valid}/{len(graphs)} valid graphs "
          f"({invalid} invalid)")

```

```
import numpy as np
import os

def graph_statistics(graphs):
    """
    Compute summary statistics over a list of (sent_id, G) graphs.
    Returns a dictionary of mean values.
    """
    sizes, densities, avg_degrees = [], [], []
    avg_path_lengths, diameters   = [], []

    for sent_id, G in graphs:
        U = G.to_undirected()
        N = G.number_of_nodes()
        if N < 2:
            continue

        sizes.append(N)

        # Density = actual edges / possible edges
        densities.append(nx.density(G))

        # Average in-degree (dependents per node)
        in_degrees = [d for _, d in G.in_degree()]
        avg_degrees.append(np.mean(in_degrees))

        if nx.is_connected(U):
            avg_path_lengths.append(
                nx.average_shortest_path_length(U)
            )
            diameters.append(nx.diameter(U))

        return {
        "mean_sentence_size":    round(np.mean(sizes), 2),
        "mean_density":          round(np.mean(densities), 4),
        "mean_avg_in_degree":    round(np.mean(avg_degrees), 4),
        "mean_avg_path_length":  round(np.mean(avg_path_lengths), 3),
        "mean_diameter":         round(np.mean(diameters), 2),
        "total_graphs":          len(graphs)
    }


results = []
for lang in ["english", "hindi", "french", "german"]:
    with open(f"graphs/{lang}_graphs.pkl", "rb") as f:
        graphs = pickle.load(f)
    stats = graph_statistics(graphs)
    stats["language"] = lang.title()
    results.append(stats)
    print(f"{lang}: {stats}")

df = pd.DataFrame(results).set_index("language")
print("\n", df.to_string())

os.makedirs("data", exist_ok=True) # Create 'data' directory if it doesn't exist
df.to_csv("data/graph_stats.csv")
print("\nSaved: data/graph_stats.csv")
```

```
import pickle
def compute_centrality_one_sentence(sent_id, G):
    """
    Compute all 3 centrality metrics for every node
    in one sentence graph.

    Returns list of dicts — one per node:
    {
      sent_id, token_id, form, upos, pos_group,
      degree_c, betweenness_c, closeness_c
    }
    """
    N = G.number_of_nodes()
    if N < 2:
        return []

    U = G.to_undirected()

    degree_c      = nx.in_degree_centrality(G)
    betweenness_c = nx.betweenness_centrality(U, normalized=True)
    closeness_c   = nx.closeness_centrality(U)

    rows = []
    for node_id, attrs in G.nodes(data=True):
        rows.append({
            "sent_id":       sent_id,
            "token_id":      node_id,
            "form":          attrs.get("form", ""),
            "upos":          attrs.get("upos", ""),
            "pos_group":     attrs.get("pos_group", "OTHER"),
            "degree_c":      round(degree_c.get(node_id, 0.0), 6),
            "betweenness_c": round(betweenness_c.get(node_id, 0.0), 6),
            "closeness_c":   round(closeness_c.get(node_id, 0.0), 6),
            "sent_size":     N
        })
    return rows
```

```
def is_valid_tree(G):
    """
    Check all 4 tree properties.
    Returns (bool, reason_string).
    """
    N = G.number_of_nodes()
    E = G.number_of_edges()

    if N < 4:
        return False, f"too small: {N} nodes"

    if E != N - 1:
        return False, f"wrong edge count: {E} edges for {N} nodes"

    roots = [n for n in G.nodes() if G.in_degree(n) == 0]
    if len(roots) != 1:
        return False, f"wrong root count: {len(roots)} roots"

    if not nx.is_weakly_connected(G):
        return False, "graph is disconnected"

    if nx.number_of_selfloops(G) > 0:
        return False, "self-loop detected"

    return True, "ok"

def filter_valid_graphs(graphs, language):
    """
    Filter graph list to keep only valid trees.
    Returns (valid_graphs, report_dict).
    """
    valid, invalid = [], []
    reasons = {}

    for sent_id, G in graphs:
        ok, reason = is_valid_tree(G)
        if ok:
            valid.append((sent_id, G))
        else:
            invalid.append((sent_id, reason))
            reasons[reason] = reasons.get(reason, 0) + 1

    print(f"{language}: {len(valid)} valid, {len(invalid)} invalid")
    if reasons:
        print(f"  Invalid reasons: {reasons}")

    return valid, {
        "language":      language,
        "total":         len(graphs),
        "valid":         len(valid),
        "invalid":       len(invalid),
        "pct_valid":     round(100 * len(valid) / len(graphs), 2)
    }
```

```
def get_sentence_length(G):
    """Return number of nodes (tokens) in graph."""
    return G.number_of_nodes()


def assign_bin(length):
    """Assign a sentence to a length bin."""
    if 4 <= length <= 8:
        return "short"
    elif 9 <= length <= 18:
        return "medium"
    else:
        return "long"


def stratified_sample(valid_graphs, n_total=2000, seed=42):
    """
    Sample n_total graphs with proportional allocation
    across sentence length bins.

    Parameters:
      valid_graphs : list of (sent_id, G) tuples
      n_total      : total number of graphs to sample
      seed         : random seed for reproducibility

    Returns:
      sampled : list of (sent_id, G) tuples
      report  : dict with bin counts
    """
    random.seed(seed)

    bins = {"short": [], "medium": [], "long": []}
    for sent_id, G in valid_graphs:
        length = get_sentence_length(G)
        b = assign_bin(length)
        bins[b].append((sent_id, G))

    total_valid = len(valid_graphs)
    allocation = {}
    for bin_name, bin_graphs in bins.items():
        proportion = len(bin_graphs) / total_valid
        allocation[bin_name] = max(1, round(proportion * n_total))

    diff = n_total - sum(allocation.values())
    allocation["medium"] += diff

    sampled = []
    bin_report = {}
    for bin_name, n_sample in allocation.items():
        available = bins[bin_name]
        n_take = min(n_sample, len(available))
        sampled_bin = random.sample(available, n_take)
        sampled.extend(sampled_bin)
        bin_report[bin_name] = {
            "available": len(available),
            "sampled":   n_take,
            "proportion": round(len(available)/total_valid, 3)
        }
        print(f"  {bin_name:6}: {len(available):5} available "
              f"→ {n_take} sampled")

    random.shuffle(sampled)
    print(f"  Total sampled: {len(sampled)}")
    return sampled, bin_report
```

```
import pickle
import networkx as nx
import pandas as pd
import numpy as np
import os
import time

os.makedirs("results", exist_ok=True)


def compute_betweenness_centrality(language):

    print(f"Computing betweenness centrality for {language}...")
    print(f"  (This may take 5-15 minutes — please wait)")
    start = time.time()

    with open(f"data/sampled_{language}.pkl", "rb") as f:
        graphs = pickle.load(f)

    all_rows = []
    n_processed = 0

    for sent_id, G in graphs:
        N = G.number_of_nodes()
        if N < 4:
            continue

        # Convert to undirected for betweenness
        U = G.to_undirected()

        # Compute normalized betweenness centrality
        # normalized=True divides by (N-1)(N-2)/2
        bc = nx.betweenness_centrality(
             U, normalized=True
        )

        for node_id, attrs in G.nodes(data=True):
            all_rows.append({
                "sent_id":   sent_id,
                "token_id":  node_id,
                "form":      attrs.get("form", ""),
                "upos":      attrs.get("upos", ""),
                "pos_group": attrs.get("pos_group", "OTHER"),
                "between_c": round(bc.get(node_id, 0.0), 6),
                "sent_size": N,
                "language":  language
            })

        n_processed += 1
        # Print progress every 500 sentences
        if n_processed % 500 == 0:
            elapsed = time.time() - start
            rate = n_processed / elapsed
            remaining = (len(graphs) - n_processed) / rate
            print(f"  {n_processed}/{len(graphs)} done "
                  f"({elapsed:.0f}s elapsed, "
                  f"~{remaining:.0f}s remaining)")

    df = pd.DataFrame(all_rows)
    total_time = time.time() - start
    print(f"  Completed in {total_time:.1f}s — "
          f"{len(df)} token rows")
    return df
```

```
import pandas as pd
import numpy as np
from scipy.stats import kruskal
import os

os.makedirs("results", exist_ok=True)

# Load sentence-level means for all 3 metrics
deg  = pd.read_csv(
    "results/degree_sentence_means.csv"
).rename(columns={"mean_degree_c":  "score"})
bet  = pd.read_csv(
    "results/between_sentence_means.csv"
).rename(columns={"mean_between_c": "score"})
clos = pd.read_csv(
    "results/close_sentence_means.csv"
).rename(columns={"mean_close_c":   "score"})

deg["metric"]  = "degree"
bet["metric"]  = "betweenness"
clos["metric"] = "closeness"

all_data = pd.concat(
    [deg, bet, clos], ignore_index=True
)
all_data = all_data[all_data["pos_group"].isin(
    ["VERB","NOUN","ADJ","ADV"]
)]

POS_GROUPS = ["VERB","NOUN","ADJ","ADV"]
records = []

for metric in ["degree","betweenness","closeness"]:
    for lang in [
        "english","hindi","french","german"
    ]:
        groups = {}
        for pos in POS_GROUPS:
            scores = all_data[
                (all_data["metric"]    == metric) &
                (all_data["language"]  == lang)   &
                (all_data["pos_group"] == pos)
            ]["score"].dropna().values
            if len(scores) >= 5:
                groups[pos] = scores

        if len(groups) < 2:
            continue

        H, p = kruskal(*groups.values())
        N = sum(len(g) for g in groups.values())
        k = len(groups)
        eta_sq = (H - k + 1) / (N - k)
        eta_sq = max(0, round(eta_sq, 4))
        if eta_sq >= 0.14:
            effect = "large"
        elif eta_sq >= 0.06:
            effect = "medium"
        else:
            effect = "small"

        sig = "YES" if p < 0.05 else "NO"

        records.append({
            "metric":     metric,
            "language":   lang,
            "H_stat":     round(H, 3),
            "p_value":    round(p, 8),
            "eta_sq":     eta_sq,
            "effect_size":effect,
            "significant":sig,
            "N":          N,
            "k":          k
        })

        print(f"{metric:<12} {lang:<8} "
              f"H={H:.2f} p={p:.2e} "
              f"η²={eta_sq:.4f} ({effect}) "
              f"sig={sig}")

df_results = pd.DataFrame(records)
df_results.to_csv(
    "results/kruskal_results.csv", index=False
)
print(f"\nSaved: results/kruskal_results.csv")
print(f"Significant: "
      f"{(df_results['significant']=='YES').sum()}"
      f"/{len(df_results)} tests")
```

```
import pandas as pd
import numpy as np
import scikit_posthocs as sp
import os

os.makedirs("results", exist_ok=True)

deg  = pd.read_csv(
    "results/degree_sentence_means.csv"
).rename(columns={"mean_degree_c":  "score"})
bet  = pd.read_csv(
    "results/between_sentence_means.csv"
).rename(columns={"mean_between_c": "score"})
clos = pd.read_csv(
    "results/close_sentence_means.csv"
).rename(columns={"mean_close_c":   "score"})

deg["metric"]  = "degree"
bet["metric"]  = "betweenness"
clos["metric"] = "closeness"

all_data = pd.concat(
    [deg, bet, clos], ignore_index=True
)
all_data = all_data[all_data["pos_group"].isin(
    ["VERB","NOUN","ADJ","ADV"]
)]

all_dunn = []

for metric in ["degree","betweenness","closeness"]:
    for lang in [
        "english","hindi","french","german"
    ]:
        subset = all_data[
            (all_data["metric"]   == metric) &
            (all_data["language"] == lang)
        ][["pos_group","score"]].dropna()
        if len(subset) < 20:
            continue

        # Dunn test with Bonferroni correction
        # p_adjust='bonferroni' multiplies p by n_comparisons
        dunn = sp.posthoc_dunn(
            subset,
            val_col="score",
            group_col="pos_group",
            p_adjust="bonferroni"
        )

        # Flatten matrix to rows
        pos_list = dunn.index.tolist()
        for i, pos1 in enumerate(pos_list):
            for j, pos2 in enumerate(pos_list):
                if j <= i:
                    continue  # upper triangle only
                p_val = dunn.loc[pos1, pos2]
                sig = p_val < 0.05
                all_dunn.append({
                    "metric":   metric,
                    "language": lang,
                    "group1":   pos1,
                    "group2":   pos2,
                    "p_adj":    round(p_val, 6),
                    "significant": sig
                })

        # Print VERB comparisons
        print(f"\n{metric} / {lang} — "
              f"VERB pairwise p-values (Bonferroni):")
        for pos in ["NOUN","ADJ","ADV"]:
            if pos in dunn.index:
                p = dunn.loc["VERB", pos]
                print(f"  VERB vs {pos}: "
                      f"p={p:.2e} "
                      f"{'*' if p<0.05 else 'ns'}")

df_dunn = pd.DataFrame(all_dunn)
df_dunn.to_csv(
        "results/dunn_results.csv", index=False
)
print(f"\nSaved: results/dunn_results.csv")
print(f"Total pairs tested: {len(df_dunn)}")
verb_pairs = df_dunn[
    (df_dunn["group1"]=="VERB") |
    (df_dunn["group2"]=="VERB")
]
print(f"VERB pairs significant: "
      f"{verb_pairs['significant'].sum()}"
      f"/{len(verb_pairs)}")
```

```
import pandas as pd
import numpy as np
from scipy.stats import f_oneway
import os

os.makedirs("results", exist_ok=True)

deg  = pd.read_csv(
    "results/degree_sentence_means.csv"
).rename(columns={"mean_degree_c":  "score"})
bet  = pd.read_csv(
    "results/between_sentence_means.csv"
).rename(columns={"mean_between_c": "score"})
clos = pd.read_csv(
    "results/close_sentence_means.csv"
).rename(columns={"mean_close_c":   "score"})

deg["metric"]  = "degree"
bet["metric"]  = "betweenness"
clos["metric"] = "closeness"

all_data = pd.concat(
    [deg, bet, clos], ignore_index=True
)
all_data = all_data[all_data["pos_group"].isin(
    ["VERB","NOUN","ADJ","ADV"]
)]

POS_GROUPS = ["VERB","NOUN","ADJ","ADV"]
records = []

for metric in ["degree","betweenness","closeness"]:
    for lang in [
        "english","hindi","french","german"
    ]:
        groups = {}
        for pos in POS_GROUPS:
            scores = all_data[
                (all_data["metric"]    == metric) &
                (all_data["language"]  == lang)   &
                                (all_data["pos_group"] == pos)
            ]["score"].dropna().values
            if len(scores) >= 5:
                groups[pos] = scores

        if len(groups) < 2:
            continue

        F, p = f_oneway(*groups.values())

        # Eta-squared for ANOVA
        # η² = SS_between / SS_total
        grand_mean = np.concatenate(
            list(groups.values())
        ).mean()
        ss_between = sum(
            len(g) * (g.mean() - grand_mean)**2
            for g in groups.values()
        )
        ss_total = sum(
            ((g - grand_mean)**2).sum()
            for g in groups.values()
        )
        eta_sq = round(ss_between / ss_total, 4)

        sig = "YES" if p < 0.05 else "NO"
        records.append({
            "metric":      metric,
            "language":    lang,
            "F_stat":      round(F, 3),
            "p_value":     round(p, 8),
            "eta_sq":      eta_sq,
            "significant": sig
                    })
        print(f"{metric:<12} {lang:<8} "
              f"F={F:.2f} p={p:.2e} "
              f"η²={eta_sq:.4f} sig={sig}")

df_anova = pd.DataFrame(records)
df_anova.to_csv(
    "results/anova_results.csv", index=False
)
print("\nSaved: results/anova_results.csv")
```

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.use("Agg")
import seaborn as sns
from scipy import stats
import os
%matplotlib inline
os.makedirs("figures", exist_ok=True)


print("Loading data...")

deg_raw  = pd.read_csv("results/degree_raw_english.csv")
bet_raw  = pd.read_csv("results/between_raw_english.csv")
clos_raw = pd.read_csv("results/close_raw_english.csv")

deg_sent  = pd.read_csv("results/degree_sentence_means.csv")
bet_sent  = pd.read_csv("results/between_sentence_means.csv")
clos_sent = pd.read_csv("results/close_sentence_means.csv")

deg_sent  = deg_sent.rename(columns={"mean_degree_c":  "score"})
bet_sent  = bet_sent.rename(columns={"mean_between_c": "score"})
clos_sent = clos_sent.rename(columns={"mean_close_c":  "score"})

deg_sent["metric"]  = "Degree"
bet_sent["metric"]  = "Betweenness"
clos_sent["metric"] = "Closeness"

all_sent = pd.concat([deg_sent, bet_sent, clos_sent], ignore_index=True)
all_sent = all_sent[all_sent["pos_group"].isin(["VERB","NOUN","ADJ","ADV"])]

print(f"  Sentence-level rows: {len(all_sent)}")

print("Plotting violin plots...")

POS_ORDER   = ["VERB", "NOUN", "ADJ", "ADV"]
POS_COLORS  = {
    "VERB":  "#F0997B",
    "NOUN":  "#85B7EB",
    "ADJ":   "#97C459",
    "ADV":   "#EF9F27"
}
METRICS     = ["Degree", "Betweenness", "Closeness"]
YLABELS     = {
    "Degree":      "Mean Degree Centrality (C_D)",
    "Betweenness": "Mean Betweenness Centrality (C_B)",
    "Closeness":   "Mean Closeness Centrality (C_C)"
}
LANGUAGES   = ["english", "hindi", "french", "german"]

fig, axes = plt.subplots(3, 4, figsize=(18, 13), sharey="row")

for row_idx, metric in enumerate(METRICS):
    metric_data = all_sent[all_sent["metric"] == metric]

    for col_idx, lang in enumerate(LANGUAGES):
        ax = axes[row_idx][col_idx]
        lang_data = metric_data[metric_data["language"] == lang]

        data_list = [
            lang_data[lang_data["pos_group"] == pos]["score"].dropna().values
            for pos in POS_ORDER
        ]

        # Skip if any group empty
        if any(len(d) == 0 for d in data_list):
            ax.text(0.5, 0.5, "No data", ha="center",
                    va="center", transform=ax.transAxes, fontsize=9)
            continue

        parts = ax.violinplot(
            data_list,
            positions=range(len(POS_ORDER)),
            showmedians=True,
            showextrema=True
        )

        for i, (body, pos) in enumerate(zip(parts["bodies"], POS_ORDER)):
            body.set_facecolor(POS_COLORS[pos])
            body.set_alpha(0.75)

        parts["cmedians"].set_color("#333333")
        parts["cbars"].set_color("#888888")
        parts["cmins"].set_color("#888888")
        parts["cmaxes"].set_color("#888888")

        ax.set_xticks(range(len(POS_ORDER)))
        ax.set_xticklabels(POS_ORDER, fontsize=9)
        ax.spines[["top", "right"]].set_visible(False)
        if row_idx == 0:
            ax.set_title(lang.title(), fontsize=12, fontweight="bold", pad=8)

        if col_idx == 0:
            ax.set_ylabel(YLABELS[metric], fontsize=9)

from matplotlib.patches import Patch
legend_handles = [
    Patch(facecolor=POS_COLORS[p], alpha=0.75, label=p)
    for p in POS_ORDER
]
fig.legend(handles=legend_handles, loc="lower center",
           ncol=4, fontsize=11, title="POS Group",
           title_fontsize=11, bbox_to_anchor=(0.5, -0.01))

fig.suptitle(
    "Distribution of centrality scores by POS group across 4 languages\n"
    "(horizontal line = median)",
    fontsize=13, fontweight="bold", y=1.01
)
plt.tight_layout()
plt.show()
plt.savefig("figures/eda_violin_plots.png",
            dpi=180, bbox_inches="tight")
plt.close()
print("  Saved: figures/eda_violin_plots.png")


print("Plotting Q-Q plots...")

fig, axes = plt.subplots(2, 4, figsize=(18, 8))

for col_idx, lang in enumerate(LANGUAGES):
    ax = axes[0][col_idx]
    data = deg_sent[
        (deg_sent["language"] == lang) &
        (deg_sent["pos_group"] == "VERB")
    ]["score"].dropna()

    if len(data) > 10:
        (osm, osr), (slope, intercept, r) = stats.probplot(
            data, dist="norm"
        )
        ax.scatter(osm, osr, color="#F0997B", s=8, alpha=0.5, zorder=3)
        xline = np.array([osm.min(), osm.max()])
        ax.plot(xline, slope * xline + intercept,
                color="#444441", linewidth=1.5, zorder=4)
        ax.text(0.05, 0.92, f"r={r:.3f}",
                transform=ax.transAxes, fontsize=8,
                color="#444441")
    else:
        ax.text(0.5, 0.5, "No data", ha="center",
                va="center", transform=ax.transAxes)

    ax.spines[["top", "right"]].set_visible(False)
    if col_idx == 0:
        ax.set_ylabel("Sample Quantiles\n(Degree C_D — VERB)", fontsize=9)
    ax.set_title(lang.title(), fontsize=11, fontweight="bold")
    ax.set_xlabel("Theoretical Quantiles", fontsize=8)

    ax = axes[1][col_idx]
    data2 = clos_sent[
        (clos_sent["language"] == lang) &
        (clos_sent["pos_group"] == "VERB")
    ]["score"].dropna()

    if len(data2) > 10:
        (osm2, osr2), (slope2, intercept2, r2) = stats.probplot(
            data2, dist="norm"
        )
        ax.scatter(osm2, osr2, color="#85B7EB", s=8, alpha=0.5, zorder=3)
        xline2 = np.array([osm2.min(), osm2.max()])
        ax.plot(xline2, slope2 * xline2 + intercept2,
                color="#444441", linewidth=1.5, zorder=4)
        ax.text(0.05, 0.92, f"r={r2:.3f}",
                transform=ax.transAxes, fontsize=8,
                color="#444441")
    else:
        ax.text(0.5, 0.5, "No data", ha="center",
                va="center", transform=ax.transAxes)

    ax.spines[["top", "right"]].set_visible(False)
    if col_idx == 0:
        ax.set_ylabel("Sample Quantiles\n(Closeness C_C — VERB)", fontsize=9)
    ax.set_xlabel("Theoretical Quantiles", fontsize=8)

fig.suptitle(
    "Q-Q plots for VERB sentence-level centrality scores\n"
    "Top row: Degree C_D (right-skewed) — Bottom row: Closeness C_C (near-normal)",
    fontsize=13, fontweight="bold"
)
plt.tight_layout()
plt.show()
plt.savefig("figures/eda_qq_plots.png",
            dpi=180, bbox_inches="tight")
plt.close()
print("  Saved: figures/eda_qq_plots.png")
print("Plotting correlation heatmaps...")

fig, axes = plt.subplots(1, 4, figsize=(18, 5))

for col_idx, lang in enumerate(LANGUAGES):
    ax = axes[col_idx]

    try:
        d = pd.read_csv(f"results/degree_raw_{lang}.csv")[
            ["sent_id", "token_id", "degree_c"]
        ]
        b = pd.read_csv(f"results/between_raw_{lang}.csv")[
            ["sent_id", "token_id", "between_c"]
        ]
        c = pd.read_csv(f"results/close_raw_{lang}.csv")[
            ["sent_id", "token_id", "close_c"]
        ]
    except FileNotFoundError as e:
        ax.text(0.5, 0.5, f"File not found:\n{e}",
                ha="center", va="center", transform=ax.transAxes,
                fontsize=8, color="red")
        continue

    merged = d.merge(b, on=["sent_id", "token_id"]).merge(
        c, on=["sent_id", "token_id"]
    )

    corr = merged[["degree_c", "between_c", "close_c"]].corr(
        method="spearman"
    ).round(3)
    corr.index   = ["Degree", "Betweenness", "Closeness"]
    corr.columns = ["Degree", "Betweenness", "Closeness"]

    mask = np.triu(np.ones_like(corr, dtype=bool), k=1)

    sns.heatmap(
        corr,
        annot=True,
        fmt=".3f",
        cmap="Blues",
        vmin=0, vmax=1,
        square=True,
        linewidths=0.5,
        linecolor="white",
        ax=ax,
        mask=False,
        cbar=(col_idx == 3),
        cbar_kws={"shrink": 0.8, "label": "Spearman r"}
    )
    ax.set_title(lang.title(), fontsize=12, fontweight="bold", pad=8)
    ax.tick_params(axis="x", rotation=30, labelsize=9)
    ax.tick_params(axis="y", rotation=0,  labelsize=9)

fig.suptitle(
    "Spearman correlation between centrality metrics — per language\n"
    "(values > 0.6 indicate strong agreement between metrics)",
    fontsize=13, fontweight="bold"
)
plt.tight_layout()
plt.show()
plt.savefig("figures/eda_correlation_heatmap.png",
            dpi=180, bbox_inches="tight")
plt.close()
print("  Saved: figures/eda_correlation_heatmap.png")
```

```
import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from scipy.stats import chi2
import warnings
warnings.filterwarnings("ignore")
# from lmm_prep import prepare_lmm_data

records = []

for metric in ["degree","betweenness","closeness"]:
    print(f"\nModel comparison for {metric}...")
    df = prepare_lmm_data(metric)

    # NULL MODEL: intercept + random effect only
    # (no POS, no language, no sent_len)
    null_model = smf.mixedlm(
        formula="score ~ 1",
        data=df,
        groups=df["sent_group"]
    )
    # Use ML (not REML) for LRT between models
    # with different fixed effects
    null_fit = null_model.fit(
        method="powell", reml=False # Changed method to 'powell'
    )

    # FULL MODEL: all fixed effects
    full_model = smf.mixedlm(
        formula=(
            "score ~ pos_group + "
            "language + sent_len_z"
        ),
                data=df,
        groups=df["sent_group"]
    )
    full_fit = full_model.fit(
        method="powell", reml=False # Changed method to 'powell'
    )

    # Likelihood Ratio Test
    # LRT statistic = 2 * (logL_full - logL_null)
    lrt_stat = 2 * (
        full_fit.llf - null_fit.llf
    )
    # Degrees of freedom = difference in parameters
    df_diff = (
        len(full_fit.fe_params) -
        len(null_fit.fe_params)
    )
    p_lrt = chi2.sf(lrt_stat, df_diff)

    print(f"  Null  AIC: {null_fit.aic:.1f}")
    print(f"  Full  AIC: {full_fit.aic:.1f}")
    print(f"  ΔAIC: "
          f"{null_fit.aic - full_fit.aic:.1f}")
    print(f"  LRT χ²({df_diff}) = {lrt_stat:.1f}, "
          f"p = {p_lrt:.2e}")

    records.append({
        "metric":       metric,
        "null_AIC":     round(null_fit.aic, 1),
        "full_AIC":     round(full_fit.aic, 1),
        "delta_AIC":    round(
            null_fit.aic - full_fit.aic, 1
        ),
        "null_BIC":     round(null_fit.bic, 1),
        "full_BIC":     round(full_fit.bic, 1),
        "delta_BIC":    round(
            full_fit.bic - null_fit.bic, 1
        ),
        "LRT_chi2":     round(lrt_stat, 1),
        "LRT_df":       df_diff,
        "LRT_p":        round(p_lrt, 8)
    })

df_compare = pd.DataFrame(records)
df_compare.to_csv(
    "results/lmm_model_fit.csv", index=False
)
print("\n=== MODEL FIT COMPARISON ===")
print(df_compare.to_string(index=False))
print("Saved: results/lmm_model_fit.csv")
```

```
%matplotlib inline

import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import os

# ✅ Correct path (adjust if needed)
df = pd.read_csv("/content/data/clean_files/clean_english.csv")
df = normalize_pos(df)

# Select sample
sample = None
for sid, grp in df.groupby("sent_id"):
    if 7 <= len(grp) <= 9 and (grp["upos"] == "VERB").any():
        sample = grp
        break

if sample is None:
    raise ValueError("No valid sentence found")

G = sentence_to_graph(sample)

# Convert to undirected for centrality
U = G.to_undirected()

# Centralities
dc = nx.degree_centrality(G)
bc = nx.betweenness_centrality(U, normalized=True)
cc = nx.closeness_centrality(U)

# Color map
color_map = {
    "VERB": "#F0997B",
    "NOUN": "#85B7EB",
    "ADJ": "#97C459",
    "ADV": "#EF9F27",
    "OTHER": "#D3D1C7"
}

colors = [
    color_map.get(G.nodes[n].get("pos_group", "OTHER"), "#D3D1C7")
    for n in G.nodes()
]

# Labels with scores
labels = {}
for n in G.nodes():
    form = G.nodes[n].get("form", "")
    labels[n] = f"{form}\nD:{dc[n]:.2f} B:{bc[n]:.2f} C:{cc[n]:.2f}"

# Plot
plt.figure(figsize=(14, 7))

# Better layout
pos = nx.kamada_kawai_layout(G)

nx.draw(
    G, pos,
    labels=labels,
    node_color=colors,
    node_size=2500,
    font_size=8,
    font_weight="bold",
    arrows=True,
    arrowsize=18,
    edge_color="#888780",
    width=1.5
)

plt.title(
    "Dependency graph with centrality scores\n(D=Degree, B=Betweenness, C=Closeness)",
    fontsize=13
)

plt.tight_layout()

# ✅ MUST SHOW
plt.show()

# Save
os.makedirs("figures", exist_ok=True)
plt.savefig("figures/fig3_graph_with_scores.png", dpi=180, bbox_inches="tight")

plt.close()

print("Saved Fig 3")
```
